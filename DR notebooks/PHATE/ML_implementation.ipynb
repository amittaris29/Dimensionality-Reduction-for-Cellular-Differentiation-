```python
import phate
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from  Data_Selection import *

file = 'df_main.csv'
```

```python

selected_data, combined_equal_data  = data_selector(7200, file)
```

```python
number_of_runs = 25
collection_of_phates = []
for i in range(number_of_runs):
    phate_op = phate.PHATE()
    data_phate = phate_op.fit_transform(selected_data)
    collection_of_phates.append(data_phate)
```

```python
print(collection_of_phates)
```

```python
# Plot each PHATE result in a separate subplot
n_cols = 5
n_rows = (number_of_runs + n_cols - 1) // n_cols

plt.figure(figsize=(4 * n_cols, 4 * n_rows))

for i, data_phate in enumerate(collection_of_phates):
    ax = plt.subplot(n_rows, n_cols, i + 1)
    ax.scatter(data_phate[:, 0], data_phate[:, 1], s=5, alpha=0.7)
    ax.set_title(f"PHATE Run {i+1}")
    ax.set_xticks([])
    ax.set_yticks([])

plt.suptitle("Multiple PHATE Runs on Same Data", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
